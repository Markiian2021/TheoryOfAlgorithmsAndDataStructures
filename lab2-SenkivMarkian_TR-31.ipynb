Завдання 1. Обернути масив (reverse the array)

```python
import sys
import time
import tracemalloc
import matplotlib.pyplot as plt
import numpy as np
from memory_profiler import profile

count_array=[]
memory_array=[]
time_array=[]

def reverse_array(arr):
    arr2=[]
    for item in reversed(arr):
        arr2.append(item)
    return arr2

def result_plot(typeArray):
    global count_array
    global time_array
    global memory_array
    print(count_array)
    print(time_array)
    print(memory_array)
    plt.plot(count_array, time_array, marker = 'o')
    plt.title(f"Time {typeArray}")
    plt.xscale('log')
    plt.show()
    plt.plot(count_array, memory_array, marker = 'o')
    plt.title(f"Memory {typeArray}")
    plt.xscale('log')
    plt.show()
    count_array=[]
    memory_array=[]
    time_array=[]
    
@profile
def print_result(countArray, typeArray):
    tracemalloc.start()
    timeStart=time.time()
    random_array=[]
    if typeArray=="int":
        random_array = np.random.randint(2*countArray,size=(countArray))
    else:
        random_array = np.random.rand(countArray)
    rev_array = reverse_array(random_array)
    timeEnd=time.time()
    tracemalloc.reset_peak()
    memory_usage = sys.getsizeof(rev_array)
    count_array.append(countArray)
    time_array.append(timeEnd - timeStart)
    memory_array.append(memory_usage)

print_result(10,"int")
print_result(100,"int")
print_result(1000, "int")
print_result(10000, "int")
result_plot("int")

print_result(10,"float")
print_result(100,"float")
print_result(1000, "float")
print_result(10000, "float")
result_plot("float")
```

Завдання 2. Сортування бульбашкою (bubble sort)

```python
import sys
import time
import tracemalloc
import matplotlib.pyplot as plt
import numpy as np
from memory_profiler import profile

count_array=[]
memory_array=[]
time_array=[]

def bubble_sort(array):
  for i in range(len(array)):
    for k in range(0, len(array) - i - 1):
      if array[k] > array[k + 1]:
        n = array[k]
        array[k] = array[k+1]
        array[k+1] = n
  return array

def result_plot(typeArray):
    global count_array
    global time_array
    global memory_array
    print(count_array)
    print(time_array)
    print(memory_array)
    plt.plot(count_array, time_array, marker = 'o')
    plt.title(f"Time {typeArray}")
    plt.xscale('log')
    plt.show()
    plt.plot(count_array, memory_array, marker = 'o')
    plt.title(f"Memory {typeArray}")
    plt.xscale('log')
    plt.show()
    count_array=[]
    memory_array=[]
    time_array=[]

@profile
def print_result(countArray, typeArray):
    tracemalloc.start()
    timeStart=time.time()
    random_array=[]
    if typeArray=="int":
        random_array = np.random.randint(2*countArray,size=(countArray))
    else:
        random_array = np.random.rand(countArray)
    sort_array = bubble_sort(random_array)
    timeEnd=time.time()
    tracemalloc.reset_peak()
    memory_usage = sys.getsizeof(sort_array)
    count_array.append(countArray)
    time_array.append(timeEnd - timeStart)
    memory_array.append(memory_usage)

print_result(10,"int")
print_result(100,"int")
print_result(1000, "int")
print_result(10000, "int")
result_plot("int")

print_result(10,"float")
print_result(100,"float")
print_result(1000, "float")
print_result(1000, "float")
result_plot("float")
```

Завдання 3. Сортування вставкою (insertion sort)

```python
import sys
import time
import tracemalloc
import matplotlib.pyplot as plt
import numpy as np
from memory_profiler import profile

count_array=[]
memory_array=[]
time_array=[]

def insertion_sort(array):
    for step in range(1, len(array)):
        key = array[step]
        j = step - 1    
        while j >= 0 and key < array[j]:
            array[j + 1] = array[j]
            j = j - 1
        array[j + 1] = key
    return array


def result_plot(typeArray):
    global count_array
    global time_array
    global memory_array
    print(count_array)
    print(time_array)
    print(memory_array)
    plt.plot(count_array, time_array, marker = 'o')
    plt.title(f"Time {typeArray}")
    plt.xscale('log')
    plt.show()
    plt.plot(count_array, memory_array, marker = 'o')
    plt.title(f"Memory {typeArray}")
    plt.xscale('log')
    plt.show()
    count_array=[]
    memory_array=[]
    time_array=[]

@profile
def print_result(countArray, typeArray):
    tracemalloc.start()
    timeStart=time.time()
    random_array=[]
    if typeArray=="int":
        random_array = np.random.randint(2*countArray,size=(countArray))
    else:
        random_array = np.random.rand(countArray)
    insert_array = insertion_sort(random_array)
    timeEnd=time.time()
    tracemalloc.reset_peak()
    memory_usage = sys.getsizeof(insert_array)
    count_array.append(countArray)
    time_array.append(timeEnd - timeStart)
    memory_array.append(memory_usage)

print_result(10,"int")
print_result(100,"int")
print_result(1000, "int")
print_result(10000, "int")
result_plot("int")

print_result(10,"float")
print_result(100,"float")
print_result(1000, "float")
print_result(10000, "float")
result_plot("float")
```

Завдання 5. Бінарний пошук (binary search)


```python
import sys
import time
import tracemalloc
import numpy as np
from memory_profiler import profile

count_array=[]
memory_array=[]
time_array=[]

def quick_sort(arr):
    if len(arr) <= 1:
        return arr
    pivot = arr[len(arr) // 2]
    left = [x for x in arr if x < pivot]
    middle = [x for x in arr if x == pivot]
    right = [x for x in arr if x > pivot]
    return quick_sort(left) + middle + quick_sort(right)

def binary_search(arr, target):
    left = 0
    right = len(arr) - 1
    
    while left <= right:
        mid = (left + right) // 2
        mid_val = arr[mid]
        
        if mid_val == target:
            return mid
        elif mid_val < target:
            left = mid + 1
        else:
            right = mid - 1
    
    return -1  # Якщо елемент не знайдено

def result_plot(typeArray):
    global count_array
    global time_array
    global memory_array
    print(count_array)
    print(time_array)
    print(memory_array)
    plt.plot(count_array, time_array, marker = 'o')
    plt.title(f"Time {typeArray}")
    plt.xscale('log')
    plt.show()
    plt.plot(count_array, memory_array, marker = 'o')
    plt.title(f"Memory {typeArray}")
    plt.xscale('log')
    plt.show()
    count_array=[]
    memory_array=[]
    time_array=[]

@profile
def print_result(countArray, typeArray, searchValue):
    tracemalloc.start()
    timeStart=time.time()
    random_array=[]
    if typeArray=="int":
        random_array = np.random.randint(2*countArray,size=(countArray))
    else:
        random_array = np.random.rand(countArray)
    sort_array = quick_sort(random_array)
    array_index=binary_search(sort_array,searchValue)
    if array_index != -1:
        print(f"Елемент {searchValue} знайдено в позиції {array_index}.")
    else:
        print(f"Елемент {searchValue} не знайдено.")
    timeEnd=time.time()
    tracemalloc.reset_peak()
    memory_usage = sys.getsizeof(sort_array)
    count_array.append(countArray)
    time_array.append(timeEnd - timeStart)
    memory_array.append(memory_usage)

print_result(10,"int", 5)
print_result(100,"int", 35)
print_result(1000, "int", 50)
print_result(10000, "int", 51)
result_plot("int")

print_result(10,"float", 0.25)
print_result(100,"float", 0.042)
print_result(1000, "float", 0.421)
print_result(10000, "float", 0.642)
result_plot("float")
```
