```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

columna = ["Time", "X", "Y", "Z"]
dfFootAccelerometer = pd.read_csv('foot-accelerometer.data.csv', names = columna)
dfFootGyroscope = pd.read_csv('foot-gyroscope.data.csv', names = columna)
dfRunAccelerometer = pd.read_csv('run-accelerometer.data.csv', names = columna)
dfRunGyroscope = pd.read_csv('run-gyroscope.data.csv', names = columna)
dfStairsAccelerometer = pd.read_csv('stairs-accelerometer.data.csv', names = columna)
dfStairsGyroscope = pd.read_csv('stairs-gyroscope.data.csv', names = columna)

flag_foot=0
flag_run=0
flag_stairs=0


def plot_data(df, xAxis = "", yAxis = "", title = ""):
    plt.plot(df[xAxis], df[yAxis])
    plt.title(title)
    plt.show()

def dataset_sec(df, title1 = ""):
    plot_data(dfFootAccelerometer, "Time", "X", title1+" X")
    plot_data(dfFootAccelerometer, "Time", "Y", title1+" Y")
    plot_data(dfFootAccelerometer, "Time", "Z", title1+" Z")

def dataset_2sec(df, title1, Accelerometer_Gyroscope):
    dataset_sec(df, title1+" "+Accelerometer_Gyroscope+" Time")
    dataset_2sec_ordinate(df, title1, "X", Accelerometer_Gyroscope)
    dataset_2sec_ordinate(df, title1, "Y", Accelerometer_Gyroscope)
    dataset_2sec_ordinate(df, title1, "Z", Accelerometer_Gyroscope)

def plotLineGraph (df1, ordinate, title1):
    plt.plot(df1["Time"], df1["Diff"], label=f"Diff {ordinate}")
    plt.plot(df1["Time"], df1["Mean"], label=f"Mean {ordinate}")
    plt.plot(df1["Time"], df1["STD"], label=f"STD {ordinate}")
    plt.legend()
    plt.title(title1)
    plt.show()

def plotHistogram (df1, ordinate, title1):
    plt.hist(df1["Diff"], label="Diff")
    plt.hist(df1["Mean"], label="Mean")
    plt.hist(df1["STD"], label="STD")
    plt.legend()
    plt.title(title1)
    plt.show()

def definition_movement(diff1, mean1, std1, ordinate1, Accelerometer_Gyroscope):
    global flag_foot
    global flag_run
    global flag_stairs
    if (Accelerometer_Gyroscope == "Accelerometer"):
        #Accelerometer
        if ordinate1=="X":
            if diff1>=9 and diff1<=12.5:
                flag_foot+=1
            if mean1>=6.5 and mean1<=8:
                flag_foot+=1
            if std1>=2.5 and std1<=3.5:
                flag_foot+=1
            if diff1>=35 and diff1<=90:
                flag_run+=1
            if mean1>=10 and mean1<=18:
                flag_run+=1
            if std1>=10 and std1<=15:
                flag_run+=1
            if diff1>=10 and diff1<=20:
                flag_stairs+=1
            if mean1>=8 and mean1<=10:
                flag_stairs+=1
            if std1>=3 and std1<=5:
                flag_stairs+=1

        elif ordinate1=="Y":
            if diff1>=5 and diff1<=12:
                flag_foot+=1
            if mean1>=-10 and mean1<=-7:
                flag_foot+=1
            if std1>=2 and std1<=3.5:
                flag_foot+=1
            if diff1>=20 and diff1<=60:
                flag_run+=1
            if mean1>=-18 and mean1<=5:
                flag_run+=1
            if std1>=8 and std1<=18:
                flag_run+=1
            if diff1>=10 and diff1<=17:
                flag_stairs+=1
            if mean1>=-8 and mean1<=-6:
                flag_stairs+=1
            if std1>=1.5 and std1<=4:
                flag_stairs+=1

        elif ordinate1=="Z":
            if diff1>=5 and diff1<=12:
                flag_foot+=1
            if mean1>=1.5 and mean1<=3.5:
                flag_foot+=1
            if std1>=1.5 and std1<=4.5:
                flag_foot+=1
            if diff1>=18 and diff1<=50:
                flag_run+=1
            if mean1>=0 and mean1<=5:
                flag_run+=1
            if std1>=5 and std1<=8:
                flag_run+=1
            if diff1>=6 and diff1<=9:
                flag_stairs+=1
            if mean1>=1 and mean1<=3:
                flag_stairs+=1
            if std1>=1 and std1<=3:
                flag_stairs+=1              
    else:
        #Gyroscope
        if ordinate1=="X":
            if diff1>=2.5 and diff1<=4:
                flag_foot+=1
            if mean1>=-0.5 and mean1<=0.5:
                flag_foot+=1
            if std1>=0.5 and std1<=1:
                flag_foot+=1
            if diff1>=7.5 and diff1<=17:
                flag_run+=1
            if mean1>=-1 and mean1<=2:
                flag_run+=1
            if std1>=2 and std1<=3:
                flag_run+=1
            if diff1>=3 and diff1<=5:
                flag_stairs+=1
            if mean1>=-0.5 and mean1<=1:
                flag_stairs+=1
            if std1>=1 and std1<=1.8:
                flag_stairs+=1
                
        elif ordinate1=="Y":
            if diff1>=3 and diff1<=6:
                flag_foot+=1
            if mean1>=-1 and mean1<=0.5:
                flag_foot+=1
            if std1>=1 and std1<=1.5:
                flag_foot+=1
            if diff1>=7 and diff1<=20:
                flag_run+=1
            if mean1>=-1 and mean1<=1:
                flag_run+=1
            if std1>=2 and std1<=3:
                flag_run+=1
            if diff1>=3 and diff1<=5:
                flag_stairs+=1
            if mean1>=-0.5 and mean1<=0.5:
                flag_stairs+=1
            if std1>=1 and std1<=1.5:
                flag_stairs+=1 
    
        elif ordinate1=="Z":
            if diff1>=5 and diff1<=9:
                flag_foot+=1
            if mean1>=-1 and mean1<=1:
                flag_foot+=1
            if std1>=2 and std1<=3:
                flag_foot+=1
            if diff1>=10 and diff1<=30:
                flag_run+=1
            if mean1>=-0.5 and mean1<=0.5:
                flag_run+=1
            if std1>=2 and std1<=6:
                flag_run+=1
            if diff1>=5 and diff1<=10:
                flag_stairs+=1
            if mean1>=-0.5 and mean1<=0.5:
                flag_stairs+=1
            if std1>=2 and std1<=3:
                flag_stairs+=1 

def compare_movement(dfAccelerometer, dfGyroscope, title1):
    global flag_foot
    global flag_run
    global flag_stairs
    flag_foot=0
    flag_run=0
    flag_stairs=0
    dataset_2sec(dfAccelerometer, title1, "Accelerometer")
    dataset_2sec(dfGyroscope, title1, "Gyroscope")
    if flag_foot > flag_run and flag_foot > flag_stairs:
        print(f"Це ходьба: Foot={flag_foot};  Run={flag_run};  Stairs={flag_stairs}")
    elif flag_run > flag_stairs:
        print(f"Це біг: Foot={flag_foot};  Run={flag_run};  Stairs={flag_stairs}")
    else:
        print(f"Це ходьба по сходах: Foot={flag_foot};  Run={flag_run};  Stairs={flag_stairs}")

def dataset_2sec_ordinate(df, title1, ordinate, Accelerometer_Gyroscope):
    df1=pd.DataFrame(columns=["Time", "Diff", "Mean", "STD"])
    i_sec=2
    i=0
    i2_min=0
    i2_max=0
    value_min=df[ordinate][0]
    value_max=df[ordinate][0]
    while i < len(df):
        if df["Time"][i] < i_sec:
            i2_max=i
            if df[ordinate][i] < value_min:
                value_min = df[ordinate][i]
            if df[ordinate][i] > value_max:
                value_max = df[ordinate][i]
        else:
            subset = df[i2_min:i2_max][ordinate]
            std_value = np.std(subset)
            mean_value = np.mean(subset)
            df1.loc[len(df1)] = [i_sec, value_max - value_min, mean_value, std_value]
            i2_min=i2_max+1
            value_min=df[ordinate][i2_min]
            value_max=df[ordinate][i2_min]
            i_sec+=2
            definition_movement(value_max - value_min, mean_value, std_value, ordinate, Accelerometer_Gyroscope)
        i+=1
    plotLineGraph(df1, ordinate, title1+" "+Accelerometer_Gyroscope+" "+ordinate)
    plotHistogram(df1, ordinate, title1+" "+Accelerometer_Gyroscope+" "+ordinate)

compare_movement(dfFootAccelerometer, dfFootGyroscope, "Foot")
compare_movement(dfRunAccelerometer, dfRunGyroscope, "Run")
compare_movement(dfStairsAccelerometer, dfStairsGyroscope, "Stairs")
```
