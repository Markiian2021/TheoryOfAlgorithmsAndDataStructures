```python
import pandas as pd

columns = ["TimesPregnant", "PlasmaGlucose", "DiastolicBP", "TricepsThickness", "SerumInsulin", "BMI", "DiabetesPedigree", "Age", "Class"]

df = pd.read_csv('pima-indians-diabetes.data.csv', names = columns)
print("Розмір датасету: ", df.shape)
if df.isnull().values.any():
    print(" датасеті є пропущені дані.")
else:
    print("Пропущених даних у датасеті немає.")

print(df.isna().sum(), "\n")
```

```python
import pandas as pd

columns = ["TimesPregnant", "PlasmaGlucose", "DiastolicBP", "TricepsThickness", "SerumInsulin", "BMI", "DiabetesPedigree", "Age", "Class"]

df= pd.read_csv('pima-indians-diabetes.data.csv', names = columns)
print("\nСереднє арифметичне для кожної колонки:\n", df.mean())
print("\nДисперсія для кожної колонки:\n",df.var())
print("\nСереднє квадратичне відхилення для кожної колонки:\n", df.std())
```

```python
import pandas as pd
import matplotlib.pyplot as plt

column_names = ["TimesPregnant", "PlasmaGlucose", "DiastolicBP", "TricepsThickness", "SerumInsulin", "BMI", "DiabetesPedigree", "Age", "Class"]
df = pd.read_csv('pima-indians-diabetes.data.csv', names = column_names)

for column in df.columns:
    plt.hist(df[column], alpha = 0.5, label = column)
plt.legend()
```
